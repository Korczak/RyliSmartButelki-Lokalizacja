```python
import cv2
import os
import shutil
import matplotlib.pyplot as plt
```

```python
img_width, img_height = 312, 416
```

```python
def resize_dataset():
    resized_images = 0

    directory = "../../dataset/butelka/"
    processed_path = os.path.join("../../dataset/butelka/train_resized/")
    if os.path.exists(processed_path):
        shutil.rmtree(processed_path)
    os.mkdir(processed_path)
    for root, subdirs, files in os.walk(directory):
        for image in files:
            img = cv2.imread(root+"/"+image, cv2.IMREAD_UNCHANGED)
            if(img.shape[0] < img.shape[1]):
                img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
            img = cv2.resize(img, (img_width, img_height), interpolation = cv2.INTER_AREA)
            cv2.imwrite(processed_path + '/' + image, img);
            resized_images = resized_images + 1
            if(resized_images % 20 == 0):
                print(resized_images)
                #input('KontynuowaÄ‡?')
                
resize_dataset()
```
